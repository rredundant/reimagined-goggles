```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

import warnings
warnings.filterwarnings("ignore")

from api_keys import api_key

from citipy import citipy

output_data_file = "output_data/cities.csv"

lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
lat_lngs = []
cities = [] 
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)
```

```python
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    if city not in cities:
        cities.append(city)
        
len(cities)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?appid="+ api_key
city_weather = []
```

```python
for city in cities:
        query_url = url + "&q=" + city
        response = requests.get(query_url).json()
    #     api=f'http://ergast.com/api/f1/{year}/drivers/{driver}/status.json'
    #     resp=requests.get(api).json().get('MRData')
        city_weather.append(response)
        number = cities.index(city) + 1
        print(f"Processing city {number} | {city}.")
```

```python
weather_df = pd.DataFrame.from_records(city_weather)
weather_df = weather_df[weather_df.base.notnull()]
```

```python
weather_df.head(3)
```

```python
# weather_df_new['Cloudiness'] = weather_df.clouds[""]
weather_df['Cloudiness'] = weather_df.clouds.map(lambda x: x.get('all'))
weather_df['Country'] = weather_df.sys.map(lambda x: x.get('country'))
weather_df['Date'] = weather_df.sys.map(lambda x: x.get('sunrise'))
weather_df['Humidity'] = weather_df.main.map(lambda x: x.get('humidity'))
weather_df['Lat'] = weather_df.coord.map(lambda x: x.get('lat'))
weather_df['Lng'] = weather_df.coord.map(lambda x: x.get('lon'))
weather_df['Max_Temp'] = weather_df.main.map(lambda x: x.get('temp_max'))
weather_df['Wind_Speed'] = weather_df.wind.map(lambda x: x.get('speed'))
```

```python
pd.set_option('display.max_colwidth', -1)
weather_df.head(3)
```

```python
weather_df_new = weather_df[['name', 'Cloudiness', 'Country', 'Date', 'Humidity', 'Lat', 'Lng','Max_Temp', 'Wind_Speed']]
weather_df_new['Max_Temp'] = weather_df_new.Max_Temp.map(lambda x: ((x*9)/5 - 459.67))
```

```python
weather_df_new.head(3)
```

```python
plt.scatter(weather_df_new.Lat, 
            weather_df_new.Max_Temp,
           alpha = .33)
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.title('City Latitude vs. Maximum Temperature')
```

```python
plt.scatter(weather_df_new.Lat, 
            weather_df_new.Humidity,
           alpha = .33)
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.title('City Latitude vs. Humidity')
```

```python
plt.scatter(weather_df_new.Lat, 
            weather_df_new.Cloudiness,
           alpha = .33)
plt.xlabel('Latitude')
plt.ylabel('Cloudiness (%)')
plt.title('City Latitude vs. Cloudiness')
```

```python
plt.scatter(weather_df_new.Lat, 
            weather_df_new.Wind_Speed,
           alpha = .33)
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.title('City Latitude vs. Wind Speed')
```

# Observations
1. The only factor that we graphed that seems to have a direct relationship with latitude is maximum temperature. Although my data skews toward the northern hemisphere, a clear arc shape is evident.
2. Interestingly, the low humidity cities seem to occur mostly in at the same distance from the equator with heavy concentrations at both negative and positive 40 degrees.
